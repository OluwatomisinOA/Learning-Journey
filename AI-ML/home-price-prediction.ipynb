```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# load the dataset
df = pd.read_csv("homeprices2.csv")
print(df)
```

```python
# handle missing values
median_bedrooms = df.bedrooms.median()
df.bedrooms = df.bedrooms.fillna(median_bedrooms)
print(df)
```

```python
# prepare the dataset
x = df[['area','bedrooms','age']]
y = df['prices']
```

```python
# create and train the model
model = LinearRegression()
model.fit(x,y)
```

```python
# value of the coefficients
print("Coefficients:", model.coef_)
```

```python
# value of the intercept
print("Intercept:", model.intercept_)
```

```python
# prediction example 1: area = 3000 sq. ft., bedrooms = 3, age = 15
new_house = pd.DataFrame({'area':[3000],'bedrooms':[3],'age':[15]})
prediction = model.predict(new_house)
print("Predicted price:", prediction)
```

```python
# prediction example 2: area = 2500 sq. ft., bedrooms = 4, age = 5
new_house = pd.DataFrame({'area':[2500],'bedrooms':[4],'age':[5]})
prediction = model.predict(new_house)
print("Predicted price:", prediction)
```

```python
# scatter plot : area vs price
plt.xlabel('Area (sq ft)')
plt.ylabel('Price (US$)')
plt.title('House Price vs Area')
plt.scatter(df['area'],df['prices'])
plt.show()
```

```python
# regression line (area only)
# train a temporary single-variable model
x_area = df[['area']]
y = df['prices']
area_model = LinearRegression()
area_model.fit(x_area, y)

plt.xlabel('Area (sq ft)')
plt.ylabel('Price')
plt.title('Linear Regression: Area vs Price')
plt.scatter(df['area'], y)
plt.plot(df['area'], area_model.predict(x_area))
plt.show()
```
