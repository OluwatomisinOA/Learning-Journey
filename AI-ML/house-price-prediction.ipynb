```python
# python code to predict house prices using simple linear regression 
```

```python
# import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
# load the data
df = pd.read_csv('homeprices.csv')
print(df)
```

```python
# plot without the regression line
%matplotlib inline
plt.xlabel('area (sq ft)')
plt.ylabel('prices (US$)')
plt.scatter(df.area, df.price, color='red', marker='+')
```

```python
# create the model
model = linear_model.LinearRegression()
```

```python
# train the data with the available data
X = df[['area']]
Y = df.price
model.fit(X, Y)
```

```python
# value of coefficient
print("Coefficient:", model.coef_)
```

```python
# value of intercept
print("Intercept:", model.intercept_)
```

```python
# prediction example
new_house_area = pd.DataFrame({'area':[3300]})
print(model.predict(new_house_area))
```

```python
# plot with the regression line
%matplotlib inline
plt.xlabel('area (sq ft)', fontsize=20)
plt.ylabel('prices (US$)', fontsize=20)
plt.scatter(df.area, df.price, color='red', marker='+')
plt.plot(df.area, model.predict(df[['area']]), color='blue')
```

```python
# using the formula y = ax = b to calculate and confirm the prediction example
pred = 135.78767123 * 3300 + 180616.43835616432
print(pred)
```

```python
# load a data with a list of areas for price prediction
d_f = pd.read_csv('areas.csv')

# print 5 values 
print(d_f.head(5))
```

```python
# make predictions
pre = model.predict(d_f)
```

```python
# store the predictions in column variable of d_f
d_f['prices'] = pre
```

```python
# convert d_f to a csv file
d_f.to_csv('prediction.csv', index=False)
print("Done!")
```

```python
import pickle
```

```python
with open('model_pickle', 'wb') as file:
    pickle.dump(model, file)
```

```python
with open('model_pickle', 'rb') as file:
    model_p = pickle.load(file) 
```

```python
# values of intercept and coefficient using pickle
print(f"Coefficient:", model_p.coef_)
print(f"Intercept:", model_p.intercept_)
```

```python
new_house_area = pd.DataFrame({'area':[5000]})
print(model_p.predict(new_house_area))
```
