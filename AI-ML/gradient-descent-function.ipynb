```python
# import libraries
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
# define the variables
arr = np.array([1,2,3,4,5])
x = arr.reshape(-1,1)
y = np.array([5,7,9,11,13])
```

```python
# create the model and fit it
model = LinearRegression()
model.fit(x,y)
```

```python
# value of coefficient (m) and intercept (b) 
print("Coefficient (m):", model.coef_)
print("Intercept(b):", model.intercept_)
```

```python
# python code to implement gradient descent for a given value of x and y
```

```python
# define the function
def gradient_descent(x,y):
    m_current = b_current = 0
    iterations = 10000
    n = len(x)
    learning_rate = 0.08
    for i in range(iterations):
        y_predicted = m_current * x + b_current
        mse = (1/n) * sum([val**2 for val in (y-y_predicted)])
        md = -(2/n) * sum(x*(y-y_predicted))
        bd = -(2/n) * sum((y-y_predicted))
        m_current -= learning_rate * md
        b_current -= learning_rate * bd
        print(f"{i+1}: m: {m_current}, b: {b_current}, mse: {mse}")
```

```python
# define the variables and call the function
x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])
gradient_descent(x,y)
```
