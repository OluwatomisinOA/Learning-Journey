```python
# given canada's adjusted net income per capita,
# python code to predict the net income in year 2020
```

```python
# import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# load the data
df = pd.read_csv('canada_per_capita_income.csv')
print(df.head(5))
```

```python
# create the model
model = LinearRegression()
```

```python
# train or fit the model with the available data
x = df[['year']]
y = df['per capita income (US$)']
model.fit(x, y)
```

```python
# make prediction income for the year 2020
year_2020 = pd.DataFrame({'year': [2020]})
prediction = model.predict(year_2020)
print("Predicted per capita income (US$) for the year 2020:", prediction)
```

```python
# scatter plot with the regression line
%matplotlib inline
plt.xlabel('Year')
plt.ylabel('Per Capita Income (US$)')
plt.title('Canada Per Capita Income vs Year')
plt.scatter(df['year'], df['per capita income (US$)'], color='blue', marker='+')
plt.plot(df['year'], model.predict(x), color='red')
```
